```python
import pandas as pd
import re
```

```python
#LECTURA EXCEL PUBLICACIONES
posts_analisis=pd.read_csv("https://raw.githubusercontent.com/jmcs1983/TFM-Sentimiento_COVID/main/Publicaciones_Analizadas.csv",sep=';')
```

```python
posts_analisis
```

```python
#LECTURA DE TODOS LOS CSV CON DATOS Y ALMACENADO EN SIN_ANALIZAR
ruta1="https://raw.githubusercontent.com/jmcs1983/TFM-Sentimiento_COVID/main/DATOS_PREVIOS/"
sin_analizar=pd.DataFrame(columns=['comentario','id_publicacion','fecha_comentario','caracteres'])
pos=0
for i in range(len(posts_analisis.index)):
  ruta2=posts_analisis.loc[i,'ARCHIVO']
  ruta=ruta1+ruta2
  post=pd.read_csv(ruta,sep=';')
  for j in range(len(post.index)):
    if post.loc[j,'object_type'] == "data":
      sin_analizar.loc[pos,'id_publicacion']=posts_analisis.loc[i,'IDENTIFICACION']
      sin_analizar.loc[pos,'comentario'] = post.loc[j,'message']
      sin_analizar.loc[pos,'fecha_comentario'] = post.loc[j,'created_time']
      pos=pos+1
```

```python
sin_analizar
```

```python
#PARA OBTENER LAS FECHAS EN FORMATO CORRECTO
from datetime import datetime
for i in range(len(sin_analizar.index)):
  sin_analizar.loc[i,'fecha_comentario']=datetime.strptime(sin_analizar.loc[i,'fecha_comentario'][0:10], '%Y-%m-%d').strftime('%Y-%m-%d')
```

```python
sin_analizar['fecha_comentario']=pd.to_datetime(sin_analizar['fecha_comentario'])
```

```python
sin_analizar
```

```python
#PROCESADO DE LOS COMENTARIOS PARA ELIMINAR: EMOJIS, URLs, HASHTAGS, MENCIONES Y ESPACIOS EN BLANCO
emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)


sin_analizar['comentario']=sin_analizar['comentario'].astype(str)
for i in range(len(sin_analizar.index)):
  sin_analizar.loc[i,'comentario']=emoji_pattern.sub(r'', sin_analizar.loc[i,'comentario']) #EMOJIS
  sin_analizar.loc[i,'comentario']=re.sub("http*\S+", " ", sin_analizar.loc[i,'comentario']) #URLS
  sin_analizar.loc[i,'comentario']=re.sub("@\S+", " ", sin_analizar.loc[i,'comentario']) #MENTIONS
  sin_analizar.loc[i,'comentario']=re.sub("#\S+", " ", sin_analizar.loc[i,'comentario']) #HASTAGS
  sin_analizar.loc[i,'comentario']=re.sub('\s{2,}', " ", sin_analizar.loc[i,'comentario']) #ESPACIOS EN BLANCO INTERNOS
  sin_analizar.loc[i,'comentario']=sin_analizar.loc[i,'comentario'].strip()
```

```python
sin_analizar
```

```python
#CALCULO DEL NUMERO DE LETRAS DE CADA COMENTARIO
for i in range(len(sin_analizar.index)):
  texto=sin_analizar.loc[i,'comentario']
  letras=0
  noletras=0
  for j in texto:
    if j.isalpha():
      letras=letras+1;
    else:
      noletras=noletras+1;
  sin_analizar.loc[i,'caracteres']=letras
```

```python
sin_analizar
```

```python
#TOTAL COMENTARIOS POR MES SIN COMPROBAR LONGITUD DE CARACTERES
sin_analizar['comentario'].groupby([sin_analizar['fecha_comentario'].dt.year.rename('a√±o'), sin_analizar['fecha_comentario'].dt.month.rename('mes')]).agg({'count'})
```

```python
sin_analizar
```

```python
#PARA CONSTRUIR DATAFRAMES DE MAYORES DE 100, MAYORES DE 75 Y MAYORES DE 50
sin_analizar['>=100'] = sin_analizar.apply(lambda x: True if x[3] > 99 else False , axis=1)
sin_analizar['>=75'] = sin_analizar.apply(lambda x: True if x[3] > 74 else False , axis=1)
sin_analizar['>=50'] = sin_analizar.apply(lambda x: True if x[3] > 49 else False , axis=1)
```

```python
sin_analizar
```

```python
#DATAFRAMES DE COMENTARIOS ANALIZADOS DE 100 O MAS CARACTERES, DE 75 O MAS CARACTERES Y DE 50 O MAS CARACTERES
analizados100 = sin_analizar[sin_analizar['>=100'] == True]
analizados50 = sin_analizar[sin_analizar['>=50'] == True]
analizados75 = sin_analizar[sin_analizar['>=75'] == True]
```

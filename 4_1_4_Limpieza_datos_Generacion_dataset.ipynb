```python
import pandas as pd
import re
```

```python
#LECTURA EXCEL PUBLICACIONES
posts_analisis=pd.read_csv("https://raw.githubusercontent.com/jmcs1983/TFM-Sentimiento_COVID/main/Publicaciones_Analizadas.csv",sep=';')
```

```python
posts_analisis
```

```python
#LECTURA DE TODOS LOS CSV CON DATOS Y ALMACENADO EN SIN_ANALIZAR
ruta1="https://raw.githubusercontent.com/jmcs1983/TFM-Sentimiento_COVID/main/DATOS_PREVIOS/"
sin_analizar=pd.DataFrame(columns=['comentario','id_publicacion','fecha_comentario','caracteres'])
pos=0
for i in range(len(posts_analisis.index)):
  ruta2=posts_analisis.loc[i,'ARCHIVO']
  ruta=ruta1+ruta2
  post=pd.read_csv(ruta,sep=';')
  for j in range(len(post.index)):
    if post.loc[j,'object_type'] == "data":
      sin_analizar.loc[pos,'id_publicacion']=posts_analisis.loc[i,'IDENTIFICACION']
      sin_analizar.loc[pos,'comentario'] = post.loc[j,'message']
      sin_analizar.loc[pos,'fecha_comentario'] = post.loc[j,'created_time']
      pos=pos+1
```

```python
sin_analizar
```

```python
#PARA OBTENER LAS FECHAS EN FORMATO CORRECTO
from datetime import datetime
for i in range(len(sin_analizar.index)):
  sin_analizar.loc[i,'fecha_comentario']=datetime.strptime(sin_analizar.loc[i,'fecha_comentario'][0:10], '%Y-%m-%d').strftime('%Y-%m-%d')
```

```python
sin_analizar['fecha_comentario']=pd.to_datetime(sin_analizar['fecha_comentario'])
```

```python
sin_analizar
```

```python
#PROCESADO DE LOS COMENTARIOS PARA ELIMINAR: EMOJIS, URLs, HASHTAGS, MENCIONES Y ESPACIOS EN BLANCO
emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  
                               u"\U0001F300-\U0001F5FF"  
                               u"\U0001F680-\U0001F6FF"  
                               u"\U0001F1E0-\U0001F1FF"  
                               u"\U00002500-\U00002BEF"  
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  
                               u"\u3030"
                               "]+", flags=re.UNICODE)


sin_analizar['comentario']=sin_analizar['comentario'].astype(str)
for i in range(len(sin_analizar.index)):
  sin_analizar.loc[i,'comentario']=emoji_pattern.sub(r'', sin_analizar.loc[i,'comentario']) #EMOJIS
  sin_analizar.loc[i,'comentario']=re.sub("http*\S+", " ", sin_analizar.loc[i,'comentario']) #URLS
  sin_analizar.loc[i,'comentario']=re.sub("@\S+", " ", sin_analizar.loc[i,'comentario']) #MENTIONS
  sin_analizar.loc[i,'comentario']=re.sub("#\S+", " ", sin_analizar.loc[i,'comentario']) #HASTAGS
  sin_analizar.loc[i,'comentario']=re.sub('\s{2,}', " ", sin_analizar.loc[i,'comentario']) #ESPACIOS EN BLANCO INTERNOS
  sin_analizar.loc[i,'comentario']=sin_analizar.loc[i,'comentario'].strip()
```

```python
sin_analizar
```

```python
#CALCULO DEL NUMERO DE LETRAS DE CADA COMENTARIO
for i in range(len(sin_analizar.index)):
  texto=sin_analizar.loc[i,'comentario']
  letras=0
  noletras=0
  for j in texto:
    if j.isalpha():
      letras=letras+1;
    else:
      noletras=noletras+1;
  sin_analizar.loc[i,'caracteres']=letras
```

```python
sin_analizar
```

```python
#COLUMNA NUEVA PARA SABER QUE CONDICIÓN CUMPLEN RESPECTO A Nº DE CARACTERES
sin_analizar['>=100'] = sin_analizar.apply(lambda x: True if x[3] > 99 else False , axis=1)
sin_analizar['>=75'] = sin_analizar.apply(lambda x: True if x[3] > 74 else False , axis=1)
sin_analizar['>=50'] = sin_analizar.apply(lambda x: True if x[3] > 49 else False , axis=1)
```

```python
#SE ELIGE LA OPCION DE 100 O MÁS CARACTERES Y SE CONSTRUYE UN NUEVO DATAFRAME
analizados100 = sin_analizar[sin_analizar['>=100'] == True]
```

```python
analizados100
```

```python
analizados100=analizados100.reset_index(drop=True)
```

```python
#FECHAS PERIODOS
fin_P1 = datetime(2020,12,25)
fin_P1=fin_P1.strftime('%Y-%m-%d')
ini_P2 = datetime(2020,12,26)
ini_P2=ini_P2.strftime('%Y-%m-%d')
fin_P2 = datetime(2021,1,17)
fin_P2=fin_P2.strftime('%Y-%m-%d')
ini_P3 = datetime(2021,1,18)
ini_P3=ini_P3.strftime('%Y-%m-%d')
fin_P3 = datetime(2021,3,11)
fin_P3=fin_P3.strftime('%Y-%m-%d')
ini_P4 = datetime(2021,3,12)
ini_P4=ini_P4.strftime('%Y-%m-%d')
fin_P4 = datetime(2021,4,12)
fin_P4=fin_P4.strftime('%Y-%m-%d')
ini_P5 = datetime(2021,4,13)
ini_P5=ini_P5.strftime('%Y-%m-%d')
fin_P5 = datetime(2021,5,16)
fin_P5=fin_P5.strftime('%Y-%m-%d')
ini_P6 = datetime(2021,5,17)
ini_P6=ini_P6.strftime('%Y-%m-%d')
fin_P6 = datetime(2021,6,8)
fin_P6=fin_P6.strftime('%Y-%m-%d')
ini_P7 = datetime(2021,6,9)
ini_P7=ini_P7.strftime('%Y-%m-%d')
fin_P7 = datetime(2021,7,15)
fin_P7=fin_P7.strftime('%Y-%m-%d')
ini_P8 = datetime(2021,7,16)
ini_P8=ini_P8.strftime('%Y-%m-%d')
```

```python
#SE AÑADE COLUMNA VACIA PARA POSTERIORMENTE IDENTIFICAR EL PERIODO
analizados100['periodo']=""
```

```python
#IDENTIFICACION DEL PERIODO DE CADA COMENTARIO
for i in range(len(analizados100.index)):
  fecha=analizados100.loc[i,'fecha_comentario'].strftime('%Y-%m-%d')
  letras=0
  noletras=0
  if fecha <= fin_P1:
    analizados100.loc[i,'periodo']=1
  elif ini_P2 <= fecha and fecha <= fin_P2:
    analizados100.loc[i,'periodo']=2
  elif ini_P3 <= fecha and fecha <= fin_P3:
    analizados100.loc[i,'periodo']=3
  elif ini_P4 <= fecha and fecha <= fin_P4:
    analizados100.loc[i,'periodo']=4
  elif ini_P5 <= fecha and fecha <= fin_P5:
    analizados100.loc[i,'periodo']=5
  elif ini_P6 <= fecha and fecha <= fin_P6:
    analizados100.loc[i,'periodo']=6
  elif ini_P7 <= fecha and fecha <= fin_P7:
    analizados100.loc[i,'periodo']=7
  else:
    analizados100.loc[i,'periodo']=8
```
